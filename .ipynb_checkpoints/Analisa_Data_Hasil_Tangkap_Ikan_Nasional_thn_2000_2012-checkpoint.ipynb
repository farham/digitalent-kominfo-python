## 2. Statistik Data

### 2.1 Jumlah Data

```python
data_produksi_ikan_rapih.count()
```

```python
jumlah_volum_ikan = data_produksi_ikan_rapih.Volume.sum()
mi = data_produksi_ikan_rapih.Volume.min()
ma = data_produksi_ikan_rapih.Volume.max()

print("Volume Ikan yang dihasilkan per Ton =", jumlah_volum_ikan)
print("Minumum =", mi)
print("Maximum =", ma)
```

### 2.2 Mean, Median

```python
import statistics #hanya untuk tipe data numerik

me = data_produksi_ikan_rapih.mean()
md = data_produksi_ikan_rapih.median()

print('Mean adalah\n', me)
print('=============')
print('Median = \n', md)
```

```python
modus_nama_ikan = data_produksi_ikan_rapih.NamaIkan.mode()
print(modus_nama_ikan)
#nama Ikan yang sering di tangkap Per Tahunnya
```

## 3. Visualisasi

### 3.1 BarChart

```python
ax = data_produksi_ikan_rapih.groupby('Tahun').mean()['Volume']
ax = ax.plot(kind='bar', title ="Rata-Rata Berdasarkan Tahun", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Volume Tangkap Ikan (Ton)", fontsize=12)
```

```python
ajp = data_produksi_ikan_rapih.groupby(['JenisPerairan']).mean()['Volume']
ajp = ajp.plot(kind='bar', title ="Rata-Rata Jenis Perairan Penangkapan Ikan", legend=True, fontsize=12)
ajp.set_xlabel("Jenis Perairan", fontsize=12)
ajp.set_ylabel("Jumlah Tangkap Ikan", fontsize=12)
```

```python
ajp = data_produksi_ikan_rapih.groupby(['JenisPerairan']).mean()['Nilai']
ajp = ajp.plot(kind='bar', title ="Rata-Rata Jenis Perairan Penangkapan Ikan", legend=True, fontsize=12)
ajp.set_xlabel("Jenis Perairan", fontsize=12)
ajp.set_ylabel("Jumlah Tangkap Ikan", fontsize=12)
```

```python
ax = data_produksi_ikan_rapih.groupby('Tahun').mean()['Nilai']
ax = ax.plot(kind='line', title ="Rata-Rata Berdasarkan Tahun", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Nilai Jual Ikan", fontsize=12)
```

### Potensi Hasil Tangkap Ikan Nasional 


Indonesia adalah negara Maritim. Sebagian besar mata pencaharian rakyat indonesia 
adalah Nelayan. Berbagai macam ikan konsumsi yang ada di indonesia baik dari perairan Laut dan Perairan Umum


Dari data yang diperoleh oleh Kementerian Kelautan dan Perikanan (KKP) pada tahun 2000 hingga 2012 adalah 63719344.191 Ton yang berhasil dihasilkan, baik dari perairan laut maupun perairan umum. Produksi terbesarnya adalah 314852.0 Ton dari satu jenis ikan per tahunnya. 

Dari 76791 Data yang didapat selama 12 Tahun oleh KKP, terdiri dari 
1. 33 Provinsi dari Aceh sampai Papua;
2. 175 Nama/Jenis Ikan;
3. 2 Jenis Perairan yaitu Laut dan Umum;
4. 12 Tahun; 
5. Volume Tangkap Ikan dengan satuan Ton;
6. Nilai Jual Ikan dalam satuan Rupiah.

Dari data diatas, dapat disimpulkan bahwa, Provinsi yang dikelilingi Laut lebih banyak menangkap Ikan dan Jenis Ikan Perairan Laut yang lebih banyak di Tangkap. Oleh karena itu jika dilihat dari Rata-Rata Jenis Perairan sebagai berikut :

```python
ajp = data_produksi_ikan_rapih.groupby(['JenisPerairan']).mean()['Volume']
ajp = ajp.plot(kind='bar', title ="Rata-Rata Jenis Perairan Penangkapan Ikan", legend=True, fontsize=12)
ajp.set_xlabel("Jenis Perairan", fontsize=12)
ajp.set_ylabel("Jumlah Tangkap Ikan", fontsize=12)
```

Jika dilihat Pertahun berdasarkan Volume Tangkap Ikan adalah 

```python
ax = data_produksi_ikan_rapih.groupby('Tahun').mean()['Volume']
ax = ax.plot(kind='line', title ="Nilai Jual Berdasarkan Tahun", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Volume Penangkapan Ikan", fontsize=12)
```

Dari tahun 2000 sampai dengan 2002 terjadi peningkatan Volume, tetapi pada tahun 2010 ke 2011 terjadi penurunan yang sangat drastis, hingga mencapai 50% penurunannya.

Jika dilihat dari Nilai Jual Ikan Pertahunnya Relativ Fluktuatif, seperti Gambar dibawah ini

```python
ax = data_produksi_ikan_rapih.groupby('Tahun').mean()['Nilai']
ax = ax.plot(kind='line', title ="Rata-Rata Berdasarkan Tahun", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Nilai Jual Ikan", fontsize=12)
```

Terjadi Peningkatan yang Signifikan dari Tahun 2005 ke 2006 hampir sekitar 80% kenaikannya, dan mengalami penurunan pada tahun 2009 ke 2010 sekitar 50%

Diharapkan dengan adanya data ini Pemerintah dapat meningkatkan Kualitas Tangkap Nasional di Indonesia dan dapat mencegah Illegal Fishing yang sedang marak pada saat ini. Dengan Meningkatnya Produksi Tangkap Nasional, Rakyat indonesia dapat menjadikan Ikan salah satu makanan yang murah lagi bergizi.
