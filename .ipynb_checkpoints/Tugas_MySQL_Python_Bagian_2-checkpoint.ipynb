```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'db_digitalent_classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

# One to Many Relationship

### 1.	Report the account representative for each customer

```python
run("""
SELECT
	a.customerName,
	a.salesRepEmployeeNumber,
	b.employeeNumber,
	CONCAT(b.firstName," ",b.lastName) as nama_sales,
	b.jobTitle 
FROM
	customers a
	INNER JOIN employees b ON a.salesRepEmployeeNumber = b.employeeNumber
""")
```

### 2.	Report total payments for "Atelier graphique".

```python
run("""
SELECT
	a.customerName,
	a.customerNumber,
	sum(b.amount) as total_bayar 
FROM
	customers a
	INNER JOIN payments b ON a.customerNumber = b.customerNumber 
WHERE
	a.customerName = 'Atelier graphique'
GROUP BY a.customerName
""")
```

### 3.	Report the total payments by date

```python
run("""
SELECT
	paymentDate,
	sum( amount ) 
FROM
	payments 
GROUP BY
	paymentDate
order by paymentDate
""")
```

### 4.	Report the products that have not been sold.

```python
run("""
SELECT
	a.productCode, 
	a.productName,
	b.quantityOrdered
FROM
	products a
left join 
	orderdetails b
on a.productCode=b.productCode
where b.quantityOrdered IS NULL

""")
```

### 5.	List the amount paid by each customer.

```python
run("""
SELECT
	a.customerNumber,
	a.customerName,
	sum( b.amount ) AS total_bayar 
FROM
	customers a
	INNER JOIN payments b ON a.customerNumber = b.customerNumber 
GROUP BY
	a.customerNumber,
	a.customerName
""")
```

### 6.	How many orders have been placed by Herkku Gifts?

```python
run("""
SELECT
	a.customerNumber,
	a.customerName,
	count( b.orderNumber ) AS total_yang_pernah_dipesan 
FROM
	customers a
	INNER JOIN orders b ON a.customerNumber = b.customerNumber 
WHERE
	a.customerName = 'Herkku Gifts' 
GROUP BY
	a.customerNumber,
	a.customerName
""")
```

### 7.	Who are the employees in Boston?

```python
run("""
SELECT
	concat( a.firstName, " ", a.lastName ) AS nama_pegawai,
	b.city 
FROM
	employees a
	INNER JOIN offices b ON a.officeCode = b.officeCode 
WHERE
	b.city = 'Boston'
""")
```

### 8.	Report those payments greater than 100,000 Dolars. Sort the report so the customer who made the highest payment appears first.

```python
run("""
SELECT
	a.customerNumber,
	b.customerName,
	a.amount 
FROM
	payments a
	INNER JOIN customers b ON a.customerNumber = b.customerNumber 
WHERE
	a.amount > 100000 
ORDER BY
	a.amount DESC
""")
```

### 9.	List the value of 'On Hold' orders.

```python
run("""
SELECT
	a.orderNumber,
	sum(quantityOrdered * priceEach) as Nilai
FROM
	orders a
	LEFT JOIN orderdetails b ON a.orderNumber=b.orderNumber
WHERE
	a.STATUS = 'On Hold'
group by a.orderNumber
""")
```

### 10.	Report the number of orders 'On Hold' for each customer.

```python
run("""
SELECT
	a.customerNumber,
	c.customerName,
	a.orderNumber,
	count( b.productCode ) AS jumlah_pesanan_on_hold 
FROM
	orders a
	LEFT JOIN orderdetails b ON a.orderNumber = b.orderNumber
	LEFT JOIN customers c ON a.customerNumber = c.customerNumber 
WHERE
	a.STATUS = 'On Hold' 
GROUP BY
	a.customerNumber,
	c.customerName,
	a.orderNumber
""")
```

# Many to Many

### 1.	List products sold by order date.

```python
run("""
SELECT
	a.orderNumber,
	a.orderDate,
	b.productCode,
	c.productName,
	a.STATUS 
FROM
	orders a
	LEFT JOIN orderdetails b ON a.orderNumber = b.orderNumber
	INNER JOIN products c ON b.productCode = c.productCode 
where 
	a.shippedDate is not null and
	a.status<>'Canceled'
ORDER BY a.orderNumber
""")
```

### 2.	List the order dates in descending order for orders for the 1940 Ford Pickup Truck.

```python
run("""
SELECT
	a.orderNumber,
	a.orderDate,
	b.productCode,
	c.productName,
	a.STATUS 
FROM
	orders a
	LEFT JOIN orderdetails b ON a.orderNumber = b.orderNumber
	INNER JOIN products c ON b.productCode = c.productCode 
where 
	c.productName='1940 Ford Pickup Truck'
ORDER BY a.orderDate desc
""")
```

### 3.	List the names of customers and their corresponding order number where a particular order from that customer has a value greater than 25,000 Dollars?

```python
run("""
SELECT
	x.customerNumber,
	y.customerName,
	x.orderNumber,
	hav.total_setiap_pemesanan 
FROM
	orders x
	INNER JOIN customers y ON x.customerNumber = y.customerNumber
	INNER JOIN (
SELECT
	a.orderNumber AS order_number,
	sum( b.quantityOrdered * b.priceEach ) AS total_setiap_pemesanan 
FROM
	orders a
	LEFT JOIN orderdetails b ON a.orderNumber = b.orderNumber 
GROUP BY
	a.orderNumber 
HAVING
	total_setiap_pemesanan > 25000 
	) AS hav ON x.orderNumber = hav.order_number
""")
```

### 4.	Are there any products that appear on all orders?

```python
run("""
SELECT 
    productCode,
    productName,
    COUNT(productCode) AS totalAppear
FROM 
    orderdetails 
        INNER JOIN
    products USING(productCode)
GROUP BY 
    productCode 
HAVING 
    totalAppear=(SELECT COUNT(*) from orders);
""")
```

### 5.	List the names of products sold at less than 80% of the MSRP.

```python
run("""
SELECT
	a.productCode,
	productName,
	MSRP,
	c.orderNumber,
	orderDate,
	STATUS,
	priceEach 
FROM
	products a
	INNER JOIN orderdetails b ON a.productCode = b.productCode
	INNER JOIN orders c ON b.orderNumber = c.orderNumber 
	AND c.STATUS = 'Shipped' 
WHERE
	priceEach < ( MSRP *.8 ) 
ORDER BY
	orderDate DESC
""")
```

### 6.	Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)

```python
run("""
    select productName, priceEach-buyPrice as keuntungan, buyPrice
    from products
    join orderdetails using (productCode)
    having keuntungan > buyPrice
""")
```

### 7.	List the products ordered on a Monday.

```python
run("""
SELECT
	a.orderNumber,
	a.orderDate,
	DAYNAME( a.orderDate ) AS hari,
	b.productCode,
	c.productName 
FROM
	orders a
	LEFT JOIN orderdetails b ON a.orderNumber = b.orderNumber
	LEFT JOIN products c ON b.productCode = c.productCode 
HAVING
	hari = 'Monday'
""")
```

### 8.	What is the quantity on hand for products listed on 'On Hold' orders?

```python
run("""
SELECT
	sum(b.quantityOrdered) as banyak_produk
FROM
	orders a
	LEFT JOIN orderdetails b ON a.orderNumber = b.orderNumber
	LEFT JOIN products c ON b.productCode = c.productCode 
WHERE
	a.STATUS = 'On Hold'
""")
```
