```python
import pymysql.cursors
import sys
import pandas as pd
import pymysql

offset = sys.argv[1]
limit = sys.argv[2]
# year = sys.argv[3]

# Connect to the database
connection = pymysql.connect(host='localhost',
                             user='root',
                             password='',
                             db='db_gdelt_fintech',
                             charset='utf8mb4',
                             cursorclass=pymysql.cursors.DictCursor)

hdr = {'User-Agent': 'Mozilla/5.0'}
```

```python
# Fungsi Untuk Running SQL
def run(sql):
    df = pd.read_sql_query(sql,connection)
    return df
```

```python
get_all_need = run("""select AvgTone, DATEADDED, SOURCEURL, Base_url, Title, 
              Isi_berita from gojek where title is not null""")
get_all_need
```

```python
get_all_need["AvgTone"] = get_all_need["AvgTone"].astype('float64')
get_all_need.isna().sum()
```

```python
get_not_null = get_all_need[(get_all_need["Title"] != '') & (get_all_need["Isi_berita"] != '')]
```

```python
get_not_null
```

```python
gb = get_not_null.groupby(['SOURCEURL','Title','Isi_berita']).mean()
gb
```

```python
gb.loc[gb['AvgTone'] > 0, 'AvgTone'] = 1
gb
```

```python
gb.loc[gb['AvgTone'] < 0, 'AvgTone'] = -1
gb
```
