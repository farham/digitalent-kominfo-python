```python
# Library Import

import pymysql
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Allow charts to apper in the notebook
%matplotlib inline
```

```python
# Function of MySQL

host = '127.0.0.1'
port = "3306"
user = 'root'
password = 'bismillah'
database = 'db_gdelt'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

### Load Data IHSG dan GDELT Foreign->ID

```python
# Load File IHSG (Monthly Based)
fileIHSG = 'IHSG2.csv'
IHSG = pd.read_csv(fileIHSG, header=0)

# Query MySQL to Pandas File GDELT (Foreign -> ID) Monthly Based
GDELT = run("""
SELECT
	t1.MonthYear,
	avg(t1.GoldsteinScale) as AvgGI,
	stddev(t1.GoldsteinScale) as StdGI,
	sum(t1.GoldsteinScale * t1.NumMentions) / count(*) as GII,
	avg(t1.AvgTone - t2.GrandMean) as Norm_AvgTone,
  sum((t1.AvgTone - t2.GrandMean) * t1.NumMentions) / count(*) as Norm_AvgToneImpact,
	stddev(t1.AvgTone) as Std_AvgTone,
	stddev(t1.AvgTone - t2.GrandMean) as Norm_Std_AvgTone,
  (t3.Sum_PositiveMentions*1.0) / sum(NumMentions) as PCTPositive,
  (t4.Sum_NegativeMentions*1.0) / sum(NumMentions) as PCTNegative,
  (t5.Sum_NeutralMentions*1.0) / sum(NumMentions) as PCTNeutral
    
FROM
	gdelt_foreign_id t1
	LEFT JOIN (
        SELECT MonthYear, avg(AvgTone) as GrandMean
        FROM gdelt_foreign_id
        GROUP BY MonthYear
					  ) t2 ON t1.MonthYear = t2.MonthYear
    LEFT JOIN (
        SELECT MonthYear, sum(NumMentions) as Sum_PositiveMentions
        FROM gdelt_foreign_id
        WHERE AvgTone > 0
				GROUP BY MonthYear
            ) t3 ON t1.MonthYear = t3.MonthYear
    LEFT JOIN (
        SELECT MonthYear, sum(NumMentions) as Sum_NegativeMentions
        FROM gdelt_foreign_id
        WHERE AvgTone < 0
				GROUP BY MonthYear
            ) t4 ON t1.MonthYear = t4.MonthYear
    LEFT JOIN (
        SELECT MonthYear, sum(NumMentions) as Sum_NeutralMentions
        FROM gdelt_foreign_id
        WHERE AvgTone = 0
				GROUP BY MonthYear
            ) t5 ON t1.MonthYear = t5.MonthYear
            
WHERE
	(t1.MonthYear >= 199706) AND (t1.MonthYear <= 201810)
    
GROUP BY
	t1.MonthYear
    
ORDER BY
	t1.MonthYear
""")
```

### ---- Description of GDELT DataFrame, All is Monthly Based ----
1. AvgGI = Average of Goldstein Index
2. StdGI = Standard Deviation of Goldstein Index
3. GII   = Goldstein Impact Index (Average Goldstein Index But Weighted By NumMentions of Events untuk memperbesar kontribusi event yang penting) ||| RUMUS, GII = sum(GI * NumMentions) / count(Jumlah Event) Tiap Bulan
5. Norm_AvgTone = Average dari AvgTone yang ternormalisasi dengan pengurangan GrandMean
6. Norm_AvgToneImpact = Schemanya sama seperti GII, cuma ini untuk AvgTone yang ternormalisasi GrandMean
7. StdAvgTone = Standard Deviation of AvgTone
8. Norm_Std_AvgTone = Standard Deviasi dari AvgTone yang ternormalisasi
9. PCTPositive, PCTNegative, PCTNeutral = Masing2 adalah persen dari berita positive, negative, dan neutral

```python
# Conditioning, Replace NaN value with 0, Karena banyak nilai null di JUMLAH BERITA NEGATIF DAN POSITIF
GDELT = GDELT.fillna(0)
GDELT
```

## Analysis Begin

#### Concat Pandas GDELT dengan IHSG

```python
data = pd.concat([GDELT,IHSG], sort='False', axis=1)
data = data.drop('Date', axis=1)
```

#### Normalize numerical data in concated pandas

```python

min_max_scaler = MinMaxScaler()
scaled_array = min_max_scaler.fit_transform(data.iloc[:,1:-2].astype('float64'))
scaled_data = data.copy()
scaled_data.iloc[:,1:-2] = scaled_array
scaled_data
```

#### Plot Scatter Matrix Untuk Lihat Korelasi Kualitatifnya

```python
pd.plotting.scatter_matrix(scaled_data.drop(columns = ['AvgGI', 'StdGI', 'Norm_AvgTone', 'Volume'], axis = 1), alpha=0.2, figsize=(20,20))
```

#### Plot Correlation Matrix Buat Liat Mana Variable Yang Berpengaruh

```python
scaled_data.drop(columns = ['AvgGI', 'StdGI', 'Norm_AvgTone', 'Volume'], axis = 1).corr()
```

### Eksplorasi Plot Antar Variable

```python
FractionMonth = scaled_data.MonthYear.str[0:4].astype('int64') + scaled_data.MonthYear.str[4:6].astype('int64') / 12

# Plot Month vs Closing Harga Saham
plt.figure(figsize = (5,5))
plt.plot(FractionMonth, scaled_data.Close)
plt.xticks(np.arange(min(FractionMonth), max(FractionMonth), 3 ))

# Plot Month vs Normalized AvgTone Impact Index
plt.figure(figsize = (5,5))
plt.plot(FractionMonth, scaled_data.Norm_AvgToneImpact)
plt.xticks(np.arange(min(FractionMonth), max(FractionMonth), 3 ))
```
