### ---- Load Data GDELT ---- <br>
### ---- Description of GDELT DataFrame, All is Day Based ----
1. AvgGI = Average of Goldstein Index
2. StdGI = Standard Deviation of Goldstein Index
3. GII   = Goldstein Impact Index (Average Goldstein Index But Weighted By NumMentions of Events untuk memperbesar kontribusi event yang penting) ||| RUMUS, GII = sum(GI * NumMentions) / count(Jumlah Event) Tiap Hari
5. Norm_AvgTone = Average dari AvgTone yang ternormalisasi dengan pengurangan GrandMean
6. Norm_AvgToneImpact = Schemanya sama seperti GII, cuma ini untuk AvgTone yang ternormalisasi GrandMean
7. StdAvgTone = Standard Deviation of AvgTone
8. Norm_Std_AvgTone = Standard Deviasi dari AvgTone yang ternormalisasi
9. PCTPositive, PCTNegative, PCTNeutral = Masing2 adalah persen dari berita positive, negative, dan neutral

```python
# Query MySQL to Pandas File GDELT (Foreign -> ID) Day Based 2018
GDELT_asli = run("""
SELECT
	t1.SQLDATE,
	#avg(t1.GoldsteinScale) as AvgGI,
	#stddev(t1.GoldsteinScale) as StdGI,
	#sum(t1.GoldsteinScale * t1.NumMentions) / count(*) as GII,
	#avg(t1.AvgTone - t2.GrandMean) as Norm_AvgTone,
    #sum((t1.AvgTone - t2.GrandMean) * t1.NumMentions) / count(*) as Norm_AvgToneImpact,
	#stddev(t1.AvgTone) as Std_AvgTone,
	#stddev(t1.AvgTone - t2.GrandMean) as Norm_Std_AvgTone,
  #(t3.Sum_PositiveMentions*100.0) / sum(NumMentions) as PCTPositive,
  #(t4.Sum_NegativeMentions*100.0) / sum(NumMentions) as PCTNegative,
  #(t5.Sum_NeutralMentions*100.0) / sum(NumMentions) as PCTNeutral
  t3.Sum_PositiveMentions as SumPositive,
  t4.Sum_NegativeMentions as SumNegative,
  t5.Sum_NeutralMentions as SumNeutral,
  sum(NumMentions) as NumMentions,
  count(*) as SumNews,
  sum(t1.AvgTone * NumMentions) as AvgToneNumMentions,
  sum(t1.AvgTone) as SumAvgTone
  
    
FROM
	gdelt_foreign_id t1
	LEFT JOIN (
        SELECT SQLDATE, avg(AvgTone) as GrandMean
        FROM gdelt_foreign_id
        GROUP BY SQLDATE
					  ) t2 ON t1.SQLDATE = t2.SQLDATE
    LEFT JOIN (
        SELECT SQLDATE, sum(NumMentions) as Sum_PositiveMentions
        FROM gdelt_foreign_id
        WHERE AvgTone > 3
				GROUP BY SQLDATE
            ) t3 ON t1.SQLDATE = t3.SQLDATE
    LEFT JOIN (
        SELECT SQLDATE, sum(NumMentions) as Sum_NegativeMentions
        FROM gdelt_foreign_id
        WHERE AvgTone < -3
				GROUP BY SQLDATE
            ) t4 ON t1.SQLDATE = t4.SQLDATE
    LEFT JOIN (
        SELECT SQLDATE, sum(NumMentions) as Sum_NeutralMentions
        FROM gdelt_foreign_id
        WHERE AvgTone BETWEEN -3 AND 3
				GROUP BY SQLDATE
            ) t5 ON t1.SQLDATE = t5.SQLDATE
            
WHERE
	t1.SQLDATE >= '20171210'
    
GROUP BY
	t1.SQLDATE
    
ORDER BY
	t1.SQLDATE
""")
```

### GDELT Conditioning

```python
# Conditioning, Replace NaN value with 0, Karena banyak nilai null di JUMLAH BERITA NEGATIF DAN POSITIF
GDELT_asli = GDELT_asli.fillna(0)

GDELT = GDELT_asli.copy()
GDELT.SQLDATE = pd.to_datetime(GDELT.SQLDATE)
GDELT.index = list(np.arange(0,len(GDELT)))
```

### Classifier Untuk Perilaku 'Saham' dari Change Persent Tone Berita 3 Hari Sebelumnya

```python
# Pilih data seperti ini
# Date      | %Positive3hari sebelum | %Negative3hari sebelum | %Neutral3hari sebelum

PCTChangePos = []
PCTChangeNeg = []
PCTChangeNet = []
PCTChangeAvgTone = []
PCTChangeAvgToneImpact = []
PCTChangeClose = []
PCTChangeOpen = []
Class = []

i=0
lagday = 2
IHSGidx = []

for day1 in IHSG.Date:
    if i!=0:
        
        # Mencari index di dataframe GDELT untuk hari i+1 dan hari i di dataframe IHSG
        idx1 = GDELT[GDELT.SQLDATE == day1].index[0]
        day0 = IHSG.Date[i-1]
        idx0 = GDELT[GDELT.SQLDATE == day0].index[0]
        
        lanjut = True
        for column in GDELT.columns[1:4]:
            lanjut = lanjut and (sum(GDELT[column][(idx0-lagday):(idx0)]) != 0)
        
        if lanjut:
            # Persentase perubahan jumlah mention berita positif
            sum0 = sum(GDELT.SumPositive[(idx0-lagday):(idx0)])
            sum1 = sum(GDELT.SumPositive[(idx1-lagday):(idx1)])
            temp = (sum1 - sum0) / sum0
            PCTChangePos.append(temp)

            # Persentase perubahan jumlah mention berita negatif
            sum0 = sum(GDELT.SumNegative[(idx0-lagday):(idx0)])
            sum1 = sum(GDELT.SumNegative[(idx1-lagday):(idx1)])
            temp = (sum1 - sum0) / sum0
            PCTChangeNeg.append(temp)

            # Persentase perubahan jumlah mention berita netral
            sum0 = sum(GDELT.SumNeutral[(idx0-lagday):(idx0)])
            sum1 = sum(GDELT.SumNeutral[(idx1-lagday):(idx1)])
            temp = (sum1 - sum0) / sum0
            PCTChangeNet.append(temp)

            # Persentase perubahan AvgTone
            sum0 = sum(GDELT.SumAvgTone[(idx0-lagday):(idx0)]) / sum(GDELT.SumNews[(idx0-lagday):(idx0)])
            sum1 = sum(GDELT.SumAvgTone[(idx1-lagday):(idx1)]) / sum(GDELT.SumNews[(idx1-lagday):(idx1)])
            temp = (sum1 - sum0) / sum0
            PCTChangeAvgTone.append(temp)

#             # Persentase perubahan AvgToneImpact
#             sum0 = sum(GDELT.AvgToneNumMentions[(idx0-lagday):(idx0)]) / sum(GDELT.SumNews[(idx0-lagday):(idx0)])
#             sum1 = sum(GDELT.AvgToneNumMentions[(idx1-lagday):(idx1)]) / sum(GDELT.SumNews[(idx1-lagday):(idx1)])
#             temp = (sum1 - sum0) / sum0
#             PCTChangeAvgToneImpact.append(temp)
            
            # Parameter perubahan saham
            PCTChangeClose.append((IHSG.Close[i] - IHSG.Close[i-1]) / IHSG.Close[i-1])
            PCTChangeOpen.append((IHSG.Open[i] - IHSG.Open[i-1]) / IHSG.Open[i-1])
            if IHSG.Close[i] < IHSG.Close[i-1]:
                Class.append('Down')
            elif IHSG.Close[i] > IHSG.Close[i-1]:
                Class.append('Up')
            IHSGidx.append(i)
            
    i = i + 1
    
    
Saham = pd.DataFrame({'Date':IHSG.Date[IHSGidx], 'PCTChangePos':PCTChangePos, 'PCTChangeNeg':PCTChangeNeg, 
                      'PCTChangeNet':PCTChangeNet, 'PCTChangeAvgTone':PCTChangeAvgTone, 
                      #'PCTChangeAvgToneImpact':PCTChangeAvgToneImpact,
                      'PCTChangeClose':PCTChangeClose, 'PCTChangeOpen':PCTChangeOpen, 'Class':Class})
Saham.index = np.arange(0,len(Saham),1)
Saham
```
