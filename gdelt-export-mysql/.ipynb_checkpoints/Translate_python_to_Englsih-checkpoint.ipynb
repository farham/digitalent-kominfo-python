```python
from textblob import TextBlob

text = '''Shiliew merupakan mantan kepala penjaga penjara Impel 
Down yang bergabung dengan Bajak Laut Kurohige. 
Sebelumnya, dia dipenjara karena kekerasan
'''

blob = TextBlob(text)
blob.tags           # [('The', 'DT'), ('titular', 'JJ'),
                    #  ('threat', 'NN'), ('of', 'IN'), ...]

blob.noun_phrases   # WordList(['titular threat', 'blob',
                    #            'ultimate movie monster',
                    #            'amoeba-like mass', ...])

for sentence in blob.sentences:
    print(sentence.sentiment.polarity)
# 0.060
# -0.341

blob.translate(to="en")  # 'La amenaza titular de The Blob...'
```
