```python
import pandas as pd

nama_produk = []
harga_produk = []

produk = soup.find_all("a","product-list__title")
for p in produk:
    nama = p.text
    nama_produk.append(nama)
    
harga = soup.find_all("span","product-list__price")

for h in harga:
    harga = h.text
    harga_produk.append(harga)

produk_dict = {'nama':nama_produk,'harga':harga_produk}
df = pd.DataFrame(produk_dict,columns = ['nama','harga'])
print(df)
df.sort_values('harga')
```
